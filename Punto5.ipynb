```python
import networkx as nx #grafos
import matplotlib.pyplot as plt #graficas
import numpy as np #matrices
import pandas as pd #dataframes
from math import inf #math es muy grande, inf nos deja toda la parte de informacion
from collections import defaultdict 
```

```python
#Definir los ejes
edges = [('2', '1'), ('2', '2'), ('0', '2'),('4', '2'),('1', '0'), ('0', '1'),('1', '3'),('1', '4'),('0', '3'),('0', '4'),('4', '4')]
```

```python
#Crear el grafo y pintarlo
grafo = nx.Graph()
grafo.add_edges_from(edges)
nx.draw(grafo, with_labels=True)
```

```python
#Adicionar el 5 que está volando
grafo.add_node('5') #añadir un nodo 1
nx.draw(grafo, with_labels=True)
```

```python
#Obtener matriz de adyasencia
A =nx.to_numpy_matrix(grafo) #numpy permite transmitir un grafo en una matriz
print(A)
```
